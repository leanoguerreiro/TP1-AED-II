```python
movies_metadata = movies_metadata.drop_nulls(subset=['adult'])
```

```python
def parse_py_dict(val):
    if val is None or val == "":
        return None
    try:
        # ast.literal_eval é mais seguro que 'eval' puro
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return None
```

```python
struct_schema = pl.Struct({
    "id": pl.Int64,
    "name": pl.Utf8,
    "poster_path": pl.Utf8,
    "backdrop_path": pl.Utf8
})
```

```python
movies_metadata = (
    movies_metadata
    .with_columns(
        pl.col("belongs_to_collection")
        .map_elements(parse_py_dict, return_dtype=struct_schema)
        # Importante: Renomear os campos AGORA para evitar que o 'id' da coleção
        # sobrescreva o 'id' do filme quando fizermos o unnest
        .struct.rename_fields([
            "collection_id",
            "collection_name",
            "collection_poster",
            "collection_backdrop"
        ])
    )
    .unnest("belongs_to_collection") # Isso remove a coluna antiga e cria as 4 novas
)
```

```python
movies_metadata
```

```python
movies_metadata = movies_metadata.drop(['budget', 'homepage', 'revenue', 'poster_path', 'collection_poster', 'collection_backdrop', 'tagline', 'video'])
```

```python
movies_metadata
```

```python
def extract_genre_names(val):
    try:
        if not val: return []
        # 1. Converte a string "[{'id': 1, 'name': 'Action'}, ...]" em lista real
        data = ast.literal_eval(val)
        # 2. Retorna apenas os valores da chave 'name'
        return [item['name'] for item in data if 'name' in item]
    except (ValueError, SyntaxError, TypeError):
        return []
```

```python
movies_metadata = movies_metadata.with_columns(
    pl.col("genres")
    .map_elements(extract_genre_names, return_dtype=pl.List(pl.Utf8))
    .alias("genres")
)
```

```python
cols_to_stringify = [
    name for name, dtype in movies_metadata.schema.items()
    if isinstance(dtype, (pl.List, pl.Struct))
]
```

```python
movies_metadata = movies_metadata.with_columns([
    pl.col(c).list.join("|") if isinstance(movies_metadata.schema[c], pl.List)
    else pl.col(c).struct.json_encode()
    for c in cols_to_stringify
])
```

```python
movies_metadata = movies_metadata.with_columns(
    pl.col("id").cast(pl.Int64, strict=False)
).drop_nulls("id")
```

```python
links_clean = (
    links_small
    .select(["movieId", "tmdbId"])
    .drop_nulls()
    .with_columns([
        pl.col("movieId").cast(pl.Int64),
        pl.col("tmdbId").cast(pl.Int64)
    ])
)
```

```python
movies_master = (
    links_clean.lazy()
    .join(
        movies_metadata.lazy(),
        left_on="tmdbId",
        right_on="id",
        how="inner"
    )
    .select([
        "movieId",      # Chave para ligar com ratings
        "title",
        "genres",
        "vote_average",
        "release_date"
    ])
)
```

```python
movies_master
```

```python
final_df = (
    ratings.lazy() # Transforma a tabela gigante em Lazy
    .join(movies_master, on="movieId", how="inner")
    .select([
        "userId",
        "movieId",
        "rating",
        "title",
        "genres",
        "vote_average",
        "release_date"
    ])
    # streaming=True permite processar blocos maiores que a memória RAM (Divide & Conquer automático)
    .collect(streaming=True)
)
```

```python
final_df
```

```python
final_df = final_df.drop_nulls(subset=['title', 'release_date', 'vote_average'])
```
